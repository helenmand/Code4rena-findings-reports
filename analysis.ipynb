```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/repository_reports.csv")
df.head()
```

```python
count = df['report_available'].value_counts()
```

```python
labels = 'Report Available', 'Report Not Available'
sizes = [count[1], count[0]]
colors = ['gold', 'lightskyblue']
explode = (0.1, 0)

plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=140)

plt.axis('equal') 
plt.show()
```
